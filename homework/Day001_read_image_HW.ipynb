思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

如果直接用 cv2 的 split 就一樣會看到三張圖都呈現單一顏色色調，因為除了提取 r 以外，他的 g, b 也補上同樣數值了！
所以要把 g, b 的 array 直接設定成 0，就可以看到三張單色的效果了

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img_path = 'data/lena.png'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
import numpy as np
b, g, r = cv2.split(img)
zeros = np.zeros(img.shape[:2], dtype = 'uint8')
```

```python
blue_img = cv2.merge([b, zeros, zeros])
blue_img = cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB)
plt.imshow(blue_img)
```

```python
green_img = cv2.merge([zeros, g, zeros])
green_img = cv2.cvtColor(green_img, cv2.COLOR_BGR2RGB)
plt.imshow(green_img)
```

```python
red_img = cv2.merge([zeros, zeros, r])
red_img = cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB)
plt.imshow(red_img)
```
